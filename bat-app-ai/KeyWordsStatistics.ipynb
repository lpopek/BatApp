### Description: 

This notebook was created to presen in simple way frequenacy of appearing key words from papers about virus spill-over in bat species. The data were collected from Chinese DBatVir database (http://www.mgc.ac.cn/DBatVir/), and collected in simple .txt file. Articles were from 2015 till 2022, becouse earlier it wasn't mandatory to make key words list for each paper.

```python
!gdown 1-z-cY-cT-mcvS3tRlBCge1GB-3ykrxZq
```

```python
PATH = "key_words.txt"
```

```python
with open(PATH) as file:
  key_words = []
  for line in file:
    key_words.extend(line.strip()[:-1].split(';'))
key_words
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(width=1920,height=1080).generate(" ".join([word for word in key_words]))
fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
